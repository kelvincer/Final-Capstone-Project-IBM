### Neighborhood Latitude/Longitude

First we install the **uszipcode** library.

> !pip install uszipcode

We are going to import the **SearchEngine** module.

```python
from uszipcode import SearchEngine
```

After importing the module, we will obtain the latitude and longitude of each neighborhood of San Francisco. For this we will search each latitude and longitude by its zip code

```python
search = SearchEngine(simple_zipcode=True)

latitude = []
longitude = []

for index, row in san_franc_df.iterrows():
    zipcode = search.by_zipcode(row[0]).to_dict()
    latitude.append(zipcode.get("lat"))
    longitude.append(zipcode.get("lng"))

san_franc_df["Latitude"] = latitude
san_franc_df["Longitude"] = longitude
san_franc_df
```

### Foursquare

Now with the help of foursquare we are going to get the venues in each neighborhood. First we configure our ID, the limit of the results and the radius.

```python
CLIENT_ID = '' # your Foursquare ID
CLIENT_SECRET = '' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 100
radius = 700
```

We are going to use the method used in classes to obtain the venues from the latitude and longitude of each neighborhood

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        json = requests.get(url).json()
        if json["response"].get('groups'):
            results = json["response"]['groups'][0]['items']
       
        
        # return only relevant information for each nearby venue
            venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Now that we have the method we use it with the data that we have from each neighborhood of San Francisco

```python
san_franc_venues = getNearbyVenues(names=san_franc_df['Neighborhood'],
                                   latitudes=san_franc_df['Latitude'],
                                   longitudes=san_franc_df['Longitude'],
                                   radius = radius)
san_franc_venues.head()
```

Now we group each neighborhood and calculate the average frequency of occurrence of each category.

```python
san_franc_grouped = san_franc_onehot.groupby('Neighborhood').mean().reset_index()
san_franc_grouped.head()
```

We are going to write an ordering function that orders the places in descending order. This is from the course.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

We are going to create a dataframe that shows the top 10 places in each neighborhood

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = san_franc_grouped['Neighborhood']

for ind in np.arange(san_franc_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(san_franc_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Cluster

Now we are going to use machine learning techniques, especially K-means. Let's first see how many cluster we are going to use with this method.

```python
from sklearn.cluster import KMeans
%matplotlib inline
from matplotlib import pyplot as plt

import matplotlib.cm as cm
import matplotlib.colors as colors

def cluster_variance(n, x):
    variances=[]
    kmeans=[]
    outputs=[]
    K=[i for i in range(1,n+1)]
    for i in range(1,n+1):
        variance=0
        model=KMeans(n_clusters=i,random_state=82,verbose=2).fit(x)
        kmeans.append(model)
        variances.append(model.inertia_)
        
    return variances,K,n
```

```python
san_franc_grouped_clustering = san_franc_grouped.drop('Neighborhood', 1)
variances,K,n=cluster_variance(10, san_franc_grouped_clustering )
```

Let's plot the curve.

```python
plt.plot(K,variances)
plt.ylabel("Inertia ( Total Distance )")
plt.xlabel("K Value")
plt.xticks([i for i in range(1,n+1)])
plt.show()
```

We see that the slope of the curve decreases as the value of k increases. From the value 4 and 5 of k the variation of the slope is slight. For our analysis we are going to choose a value of k equal to 5.

We are going to execute k-means with 5 clusters.

```python
kclusters = 5

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(san_franc_grouped_clustering)

kmeans.labels_
```

Let's add clustering labels to dataframe and merge to add latitude/longitude to each neighborhood.

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

san_franc_merged = san_franc_df

san_franc_merged = san_franc_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

san_franc_merged
```

We will show the result obtained.

```python
map_clusters = folium.Map(location=[san_franc_lat, san_franc_long], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(san_franc_merged['Latitude'], san_franc_merged['Longitude'], san_franc_merged['Neighborhood'], san_franc_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
